```python
import vk_requests    # работа с API
import time           # приостановка работы кода
import pandas as pd   # работа с таблицами
```

```python
# Авторизация в API
token = '' # сервисный токен
api = vk_requests.create_api(service_token=token)
```

```python
# Функция для получения участников группы
def get_group_members(group_id):
    results = []
    member_count = api.groups.getById(group_id=group_id, fields=["members_count"])[0]["members_count"]

    for i in range(0, member_count, 100):
        if i%500 == 0:
            print('Шаг', int(i/100))
        member_ids = api.groups.getMembers(group_id=group_id, count=100, offset=i)['items']
        results.extend(api.users.get(user_ids=member_ids, fields=["sex", "country", "occupation", "city"]))
        time.sleep(0.33)

    print('Всего участников в ' + str(group_id) + ': ', len(results))
    return results
```

```python
# группа Воздух 2112 мюзикл-антиутопия
vozdyx2112_members = get_group_members("vozdyx2112")

# группа «Икар» расправляет крылья
ikar_members = get_group_members("rockoperaikar")
```

```python
vozdyx2112_table = pd.json_normalize(vozdyx2112_members)
vozdyx2112_table.head(10)
```

```python
ikar_table = pd.json_normalize(ikar_members)
ikar_table.head(10)
```

```python
import seaborn as sns
```

```python
df_ikar = ikar_table.copy(deep=True)
df_ikar["group"] = "ikar"
df_vozdyx2112 = vozdyx2112_table.copy(deep=True)
df_vozdyx2112["group"] = "vozdyx2112"

df = pd.concat([df_ikar, df_vozdyx2112])
```

```python
countries = df["country.title"].unique()
popular_countries = []
for country in countries:
    count = len(df[df['country.title']==country])
    if (count > 10 and count < 7000):
        popular_countries.append(country)
        print(str(country) + ": " + str(count))
```

```python
sns.set(rc={'figure.figsize':(15,9)})
plot = sns.countplot(y="group", hue="country.title", data=df[df["country.title"].isin(popular_countries)])
plot.set_title('Group members by country', fontdict={'fontsize':15}, pad=12)
```

```python
sns.set(rc={'figure.figsize':(15,9)})
plot = sns.countplot(x="group", hue="sex", data=df)
plot.set_title('Group members by sex', fontdict={'fontsize':15}, pad=12)
```

```python
sns.set(rc={'figure.figsize':(15,9)})
plot = sns.countplot(x="group", hue="occupation.type", data=df)
plot.set_title('Group members by occupation.type', fontdict={'fontsize':15}, pad=12)
```

```python
def get_popular_cities(group, min_number, max_number):
    popular_cities = []
    for city in cities:
        count = len(df[(df['city.title'] == city) & (df['group'] == group)])
        if (count > min_number and count < max_number):
            popular_cities.append(city)
    return popular_cities
```

```python
def draw_graph(group, color, min_number, max_number):
    df_russia = df[df['country.title'] == "Russia"].copy(deep=True)
    
    cities = get_popular_cities(group, min_number, max_number)
    df_group = df_russia[(df_russia['group'] == group) & (df_russia['city.title'].isin(cities))].copy(deep=True)
    df_group.dropna(subset = ['city.title'], inplace=True)
    
    dist = {group: {}}    
    G=nx.from_pandas_edgelist(df_group, 'city.title', "group", edge_attr=True)
    for node in G:
        dist[group].update({node: 1/(len(df_group[df_group['city.title']==node]) + 1)})
    
    pos = nx.kamada_kawai_layout(G, dist=dist)
    nx.draw(G, with_labels=True, font_weight='bold', pos=pos, node_color=color, edge_color="lightgrey")
```

```python
draw_graph("ikar", "lightgreen", 20, 500)
```

```python
draw_graph("vozdyx2112", "lightblue", 5, 100)
```
